This thematic issue of the Journal of Digital History focuses on different digital tools that are used to study, explore, and narrate the past. Our idea has been to collect a broad range of articles that demonstrate and critique new ways of exploring the past through computational means—all in order to open up and problematize the use of digital methods in historical research, including analytical reflections on the possibilities (and limitations) of the analytical tools in question. The photograph above is iconic within the history of philosophy. It was taken in 1923 and exists in multiple versions online—usually with a caption explaining that a young Hans-Georg Gadamer is sawing wood with Martin Heidegger at his hut in Todtnauberg. The work is demanding, and they seem to be using a wooden hand saw, a typical tool useful in the Black Forest (Schwarzwald) region. Tools would also feature in Heidegger’s seminal book Sein und Zeit, published four years later. His term Zeug (gear or tool) extended beyond its ordinary use for technical purposes, described with enigmatic and hard-to-translate German expressions like vorhanden (being on hand) and zuhanden (being at hand). Heidegger’s tools embraced all things of which we can say that they are for something, philosopher Richard Schmitt once stated in a commentary. Zeug did not only include tools, implements, utensils, instruments, and equipment, but also ”materials, stuff, ingredients, raw-materials, [and] media” (<cite id="2kvi9"><a href="#zotero%7C8918850%2FAEMKUNGF">(Schmitt, 1965)</a></cite>).

Another important notion of Heidegger is Gestell (enframing), ChatGPT reminds us (using German notions)—again hinting at the way in which technology reveals the world to us, primarily articulated in his text, Die Frage nach der Technik (1954). ”Heidegger’s approach to technology diverges from traditional philosophical views that often regard technology as a neutral tool or instrument. If Heidegger teaches us anything, it is a profound confession that technology and tools are far from impartial. ChatGPT and a wooden hand saw have many things in common. This thematic issue of the Journal of Digital History focuses on digital tools used to study, explore, and narrate the past. As is well known, within the field of digital history, there exists a rapidly increasing range of tools to write and navigate history. Online, a number of websites feature lists of DH-tools, using applications such as Jupyter notebooks, Zotero, and Scalar—as well as a wide range of data visualization scripts, geographic mapping software, and timeline tools. Furthermore, digital historians are developing specialized digital toolboxes that draw inspiration from major platform strategies for managing digital collections and explore techniques for analyzing digital source materials using artificial intelligence and machine learning. Our idea behind the thematic issue was to collect a broad range of publications that demonstrate and critique new ways of exploring the past through computational means. Our intention was to open up and problematize the use of digital methods in historical research, including analytical reflections on the possibilities—and limitations—of analytical tools used. We have hence tried to include articles that combine multiple media modalities—from text and sound, to 3D and moving images. And even if Heidegger is not present in the thematic issue, most contributions provide critical reflections on the role of digital tools in producing new (historic) knowledge.

Three articles are published within the present, and first part, of the thematic issue of the *Journal of Digital History*. In the next few weeks and months, another nine articles will be progressively published online—among them our own text on tracking and tracing audiovisual reuse, including a description of the so-called Video Reuse Detector. Dario Rodighiero (University of Groningen) does not depart from such a tool, but rather from an exhibition in his article, “Super-Vision: Tracing EPFL History Through 8,000 Doctoral Theses”. The Super-Vision project investigates the practice of academic advising by visualizing eight thousand theses at the intersection of art, science, and technology. Rodighiero presents a diachronic mapping that uses AI to shed light on a major institutional dataset that would be unobservable otherwise. All theses are analyzed with natural language processing and mapped with techniques of dimensionality reduction, combining language and time within in an interactive visualization. The title—Super-Vision—naturally has a double meaning: on the one hand, it refers to the educational practice that connects doctoral students to supervisors; on the other hand, it employs information design like a macroscope to grasp complex phenomena from a distant standpoint.

A vast amount of text is also featured in the article, “The Text Analysis Prototype for Galileo’s Library and Letters Online: GaLiLeO” by Crystal Hall (Bowdoin College). The article describes a prototype digital laboratory for text analysis and discovery, aimed to provide interpretative pathways through historical materials related to the astronomer and engineer Galileo Galilei. Importantly, the tool, named GaLiLeO, prioritizes relationships not typically captured in database design for archival material. Instead, it is inspired by a common epistemological practice in Galileo’s lifetime: the juxtaposition of objects. New knowledge arises from seeing a document in comparison to others, not solely from the inherited value attached to an author or date. Hence, the preliminary text analysis tools in GaLiLeO represent a prototype to deprioritize post-hoc organizational structures in order to emphasize relationships across documents that may have resisted or otherwise been excluded from such structures. A similar focus on database design and organization is present in the article by Kalani Craig (Indiana University Bloomington), entitled: “Designing Our Digital Past: 

Anchoring Digital-History Tool Development in the Historical Method Through Design-Based History Research”, Craig argues that many of the tools that digital humanists use are developed by non-historians, and only at a later stage adapted for particular historical questions. Consequently, the article introduces Design-Based History Research as a methodological bridge between the practices of digital-history tool design, the use of digital methods to create historical argumentation, and social-science-inspired methodological innovation. Design-Based Research is, in short, an educational-research approach to studying learning theory that integrates theory building into the process of designing new environments, as well as the ongoing evaluation and refinement of those environments.

If text is key to the authors Craig, Hall and Rodighiero, the thematic issue also includes two articles focusing on other modalities: sound and 3D. Johan Malmstedt (Umea University) has entitled his article: “Navigating the Radio Archive: Segmentation and analysis of Swedish broadcasting data”. Using computational methods, his study investigates the sonic content of public service radio alterations during the introduction of commercial broadcasting in the late 1980s and early 1990s. By analyzing 1,600 hours of radio data, the article reveals a shift in the number of detected occurrences, with varying sonic sequences that reflect the overall structure of the broadcast. Malmstedt makes a fine contribution to understanding historical radio data, not the least by offering preprocessing and segmentation methods for working with sound recordings. He also emphasizes the methodological implications of combining dimensionality reduction and object classification approaches, demonstrating the value of using pre-trained and untrained algorithms for a comprehensive understanding of the local and fine-grained aspects of audio data. Finally, the thematic issue also presents a trip in 3D, the article by Susan Schreibman and Kelly Gillikin-Schoueri (both at Maastricht University), “Voyages in 3D: Creating a Multimodal Narrative of the Battle of Mount Street Bridge”. Their text departs from the research project, Contested Memories: The Battle of Mount Street Bridge, which utilizes 3D technologies to explore one of the key battles of the Easter 1916 Rising in Dublin. How many casualties did the British troops sent to quell the insurrection suffer during the battle? Using new tools and software, the project has been able to display the events in a novel way. By deploying Voyager Story—an open-source 3D viewer that combines text, audio, moving images, and 3D modeling to produce a rich environment for creating (spatial) narratives—multiple guided tours can be created. Within these, camera movements, 3D scene changes, annotation labels, and articles are organized for designing engagement between the model and the accompanying narrative.
